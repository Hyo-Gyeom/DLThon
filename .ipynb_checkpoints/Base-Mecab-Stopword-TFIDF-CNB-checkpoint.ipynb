# 데이터 불러오기

```python
import os
import pandas as pd

train_data_path ="data/train.csv"
train_data = pd.read_csv(train_data_path)
train_data.head()
```

## 데이터 전처리

```python
train_data.iloc[29,2]
```

```python
import re

def preprocess_document(document):
    document = re.sub('[^ㄱ-ㅎ가-힣a-zA-Z0-9]', ' ', document)
    document = re.sub(' +', ' ', document)
    document = document.strip()
    
    return document


preprocess_document(train_data.iloc[29,2])
```

```python
from konlpy.tag import Mecab

mecab = Mecab()

document = preprocess_document(train_data.iloc[29,2])
mecab.morphs(document)
```

```python
class_names = ['협박 대화', '갈취 대화', '직장 내 괴롭힘 대화', '기타 괴롭힘 대화']

train_data = train_data[['class', 'conversation']]
train_data['class'] = train_data['class'].map(lambda c: class_names.index(c))

train_data.head()
```

## 중복 단어 검사 및 stopword 결정

```python
train_data_by_class = [train_data[train_data['class'] == i] for i in range(4)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

vocabs = []

stop_words=[]

for data in train_data_by_class:
    vectorizer = CountVectorizer(preprocessor=preprocess_document, 
                                 tokenizer=mecab.morphs,
                                 stop_words=stop_words
                                )
    counts_by_document = vectorizer.fit_transform(data['conversation'])
    counts = np.array(counts_by_document.toarray()).sum(axis=0)
    vocab = vectorizer.vocabulary_
    vocab = sorted(vocab, key=lambda x: vocab[x])
    vocabs.append(pd.DataFrame({'word': vocab, 'count': counts}))
```

```python
words = []
for i in range(4):
    print('class', i, ':', end='')
    words_rank = vocabs[i].sort_values('count', ascending=False).iloc[:30]
    row = []
    for i, (word, count) in words_rank.iterrows():
        print(word, end=' ')
        row.append(word)
        #print(f'{word}({count})', end=' ')
    words.append(row)
    print('')
```

```python
words = list(map(set, words))
print(words[0] & words[1] & words[2] & words[3])
```

## 벡터화

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer(preprocessor=preprocess_document, 
                         tokenizer=mecab.morphs,
                         stop_words=['하', '아니', '해', '이', '네', '지', '내', '도', '가', '안', '은', '있', '아', '는', '어', '말', '게', '야', '고', '거', '에', '면', '나'])
print(vector.fit_transform(train_data['conversation']).toarray())
print(vector.vocabulary_)
```

## 데이터셋 분리

```python
from sklearn.model_selection import train_test_split

x_train = vector.transform(train_data['conversation'])
y_train = train_data['class']

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2)
```

```python
print('train size:', x_train.shape[0])
print('validation size:', x_val.shape[0])
```

## 모델 생성 및 학습

```python
from sklearn.naive_bayes import ComplementNB

model = ComplementNB()
model.fit(x_train, y_train)
```

## 모델 평가

```python
from sklearn.metrics import classification_report

print(classification_report(y_train, model.predict(x_train), zero_division=0))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def graph_confusion_matrix(model, x_test, y_test, classes_name):
    df_cm = pd.DataFrame(confusion_matrix(y_test, model.predict(x_test)), index=classes_name, columns=classes_name)
    fig = plt.figure(figsize=(12,12))
    heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=12)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=12)
    plt.ylabel('label')
    plt.xlabel('predicted value')
    plt.show()
    
graph_confusion_matrix(model, x_train, y_train, class_names)
```

```python
print(classification_report(y_val, model.predict(x_val), zero_division=0))
```
